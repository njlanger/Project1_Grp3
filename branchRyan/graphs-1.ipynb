```python
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from math import floor
from pprint import pprint
```

```python
fred_file = Path("../output_data/fred_data.csv")
fred_data = pd.read_csv(fred_file)
fred_data
```

```python
#Plotting the avg. 30 year mortgage rate against the avg. employment rate over time
y_values1 = fred_data['Avg. 30 Year Fixed Mortgage Rate']
y_values2 = fred_data['Avg. Unemployment Rate']
years = fred_data['Year']
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.plot(years, y_values1, label='Mortgage Rate', color='orange')
ax1.set_ylabel('Mortgage Rate', color='orange')
ax1.set_xlabel('Year')
ax1.tick_params('y')
ax2 = ax1.twinx()
ax2.plot(years, y_values2, label='Unemployment Rate', color='blue')
ax2.set_ylabel('Unemployment Rate', color='blue')
ax2.set_xlabel('Year')
ax2.tick_params('x')
ax1.legend()
ax2.legend()
plt.title('Avg. 30 Year Fixed Mortgage Rate vs. Avg. Unemployment Rate')
plt.show()
```

```python
#Plotting the avg. 30 year mortgage rate against the avg. Consumer Price Index
y_values1 = fred_data['Avg. 30 Year Fixed Mortgage Rate']
y_values3 = fred_data['Avg. CPI']
years = fred_data['Year']
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.plot(years, y_values1, label='Mortgage Rate', color='green')
ax1.set_ylabel('Mortgage Rate', color='green')
ax1.set_xlabel('Year')
ax1.tick_params('y')
ax2 = ax1.twinx()
ax2.plot(years, y_values3, label='Avg. CPI', color='purple')
ax2.set_ylabel('Avg. CPI', color='purple')
ax2.set_xlabel('Year')
ax2.tick_params('x')
ax1.legend()
ax2.legend()
plt.title('Avg. 30 Year Fixed Mortgage Rate vs. Avg. Consumer Price Index')
plt.show()
```
