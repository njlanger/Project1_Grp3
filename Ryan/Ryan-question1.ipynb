```python
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from math import floor
from pprint import pprint
```

```python
fred_file = Path("../output_data/fred_data.csv")
fred_data = pd.read_csv(fred_file)
fred_data
```

```python
#Plotting the avg. 30 year mortgage rate against the avg. employment rate over time
y_values1 = fred_data['Avg. 30 Year Fixed Mortgage Rate']
y_values2 = fred_data['Avg. Unemployment Rate']
years = fred_data['Year']
plt.plot(years, y_values1, label='Mortgage Rate')
plt.plot(years, y_values2, label='Unemployment Rate')
plt.legend()
plt.title('Avg. 30 Year Fixed Mortgage Rate vs. Avg. Unemployment Rate')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.show()
```
